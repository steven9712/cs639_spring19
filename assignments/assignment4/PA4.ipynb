```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier as DT

enc = OrdinalEncoder()
```

```python

traindata = pd.read_csv("train.csv")
traindata.head()
traindata.Id = pd.to_numeric(traindata.Id)
```

```python

X_train = traindata.loc[:, traindata.columns != 'class']
X_train = X_train.drop('Id', axis = 1)
X_train.head()
```

```python

Y_train = traindata['class'].to_frame()
Y_train.head()
```

```python

X_test = pd.read_csv("test.csv")
X_test.Id = pd.to_numeric(X_test.Id)
ids = X_test['Id']
X_test = X_test.drop('Id', axis = 1)
columns = X_test.columns

X_test.head()
```

```python
for attr in columns:
    X_train[attr] = enc.fit_transform(X_train[attr].values.reshape(-1, 1))
    X_test[attr] = enc.fit_transform(X_test[attr].values.reshape(-1, 1))
```

```python
classifier = DT(criterion = 'entropy', random_state = 42)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
X_test.columns.values
```

```python
submission = pd.DataFrame({
    "Id" : ids,
    "class" : Y_pred
})
submission.to_csv("mushroom_final.csv", index = False, encoding = 'utf-8')
```
